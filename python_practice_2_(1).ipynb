19. Створіть дві множини. Продемонстуйте над ними операції: обʼєднання, різниці, пересічі та симетричної різниці. Використовуйте методи, що не змінюють множини, а створюють нові.

```python
set1={0,1,3,5,7,9,10}
set2={0,2,4,6,8,10}
set3_union=set1.union(set2)
print(set3_union)
set4_dif=set1.difference(set2)
print(set4_dif)
set5_intersection=set1.intersection(set2)
print(set5_intersection)
set6_symmetric_dif=set1.symmetric_difference(set2)
print(set6_symmetric_dif)
#або через оператори "|" -- об'єднання; "&" -- перетин; "-" -- різниця;
# "^" -- симетрична різниця
```
