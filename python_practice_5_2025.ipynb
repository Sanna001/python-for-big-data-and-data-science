6. (11б). Створіть 2 дочірніх класи від 3d-фігура: Циліндр та Куб, які мають атрибути, необхідні для розрахунку об'єму. Визначте метод get_volume в кожному з них так, щоби вони розраховували об'єм у залежності від формули для кожного типу фігури. Створіть обʼєкт класу Циліндр та обʼєкт класу Куб. Виконайте операцію суми за допомогою оператора + між ними, аби знайти сумарний об'єм.

```python
class Cube(Figure3D):
  def __init__(self,edge):
    self.edge=edge
  def get_volume(self):
    return self.edge**3

class Cylinder(Figure3D):
  def __init__(self,radius,height,pi=3.14):
    self.radius=radius
    self.height=height
    self.pi=3.14
  def get_volume(self):
    return self.radius**2 * self.height * self.pi


cube_figure = Cube(5)
cylinder_figure = Cylinder(2,3)

print(f"Volume of a cube is {cube_figure.get_volume()}")
print(f"Volume of a cylinder is {cylinder_figure.get_volume()}")
print(f"Sum of volume cube and cylinder {cube_figure.__add__(cylinder_figure)}")
```

9. (11б). Створіть клас ShoppingCart, який містить список об'єктів типу Item. Реалізуйте методи для додавання айтемів, їх видалення та відображення списку айтемів.

```python
class Item:
  def __init__(self,name,price):
    self.name=name
    self.price=price

  def __repr__(self):
    return f"{self.name} ({self.price} UAH)"

class ShoppingCart:
  def __init__(self):
    self.__items=[]

  def __add__(self,item):
    self.__items.append(item)
    print(f"Item {item} added to list of items. Current list of items {self.__items}")

  def delete_item(self,item_name):
    for item in self.__items:
      if item.name == item_name:
        self.__items.remove(item)
        print(f"Item {item} deleted from list of items. Current list of items {self.__items}")
        item_found=True
        break

    if not item_found:
        print(f"Item {item_name} not found at list")

  def show_items(self):
    if not self.__items:
      print("List is empty ")
    else:
      print(f"Your list of items: {self.__items}")

item1=Item("Vintage bag", 5600)
item2=Item("Nike shoes", 4550)
item3=Item("Kiko lipgloss", 600)

my_cart=ShoppingCart()
my_cart.__add__(item1)
my_cart.__add__(item2)
my_cart.__add__(item3)
my_cart.delete_item("Nike shoes")
my_cart.show_items()
```

13. (11б). Створіть клас, що має визначений Iterator Protocol та при кожному виклику next повертає назву дня тижня, поки вони не вичерпаються. Зверніть увагу, що порядок має значення.

```python
class Week_days:
    def __init__(self):
        self.days_in_week = ["Monday", "Tuesday", "Wednesday", "Thursday", "Friday", "Saturday", "Sunday"]
        self.index = 0

    def __iter__(self):
        return self

    def __next__(self):
        if self.index < len(self.days_in_week):
            day = self.days_in_week[self.index]
            self.index += 1
            return day
        else:
            raise StopIteration

week = Week_days()
for day in week:
    print(day)
```
