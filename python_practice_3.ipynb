–Ø–∫—â–æ –≤ —ñ–º–µ–Ω—ñ —î –ª—ñ—Ç–µ—Ä–∞ "–∞"
 –∞–±–æ "i", –í–∏–≤–µ–¥—ñ—Ç—å –Ω–∞ –µ–∫—Ä–∞–Ω –ø–æ–≤—ñ–¥–æ–º–ª–µ–Ω–Ω—è "A or I is in the name".



–Ø–∫—â–æ –≤ —ñ–º–µ–Ω—ñ —î –ª—ñ—Ç–µ—Ä–∏ "o" —Ç–∞ "n", –í–∏–≤–µ–¥—ñ—Ç—å –Ω–∞ –µ–∫—Ä–∞–Ω –ø–æ–≤—ñ–¥–æ–º–ª–µ–Ω–Ω—è "O and N is in the name".

–¢–∞–∫–æ–∂ —è–∫—â–æ –≤ —ñ–º–µ–Ω—ñ –Ω–µ–º–∞—î –ª—ñ—Ç–µ—Ä–∏ "–æ", –∞–ª–µ —î "n", –í–∏–≤–µ–¥—ñ—Ç—å –Ω–∞ –µ–∫—Ä–∞–Ω –ø–æ–≤—ñ–¥–æ–º–ª–µ–Ω–Ω—è "Only N is in the name".

–í —Ä–µ—à—Ç—ñ –≤–∏–ø–∞–¥–∫—ñ–≤ –≤–∏–≤–µ–¥—ñ—Ç—å –Ω–∞ –µ–∫—Ä–∞–Ω –ø–æ–≤—ñ–¥–æ–º–ª–µ–Ω–Ω—è "There is no N in the name, but there might be O".

–ó–≤–µ—Ä–Ω—ñ—Ç—å —É–≤–∞–≥—É, —â–æ, –Ω–∞–ø—Ä–∏–∫–ª–∞–¥, –¥–ª—è —ñ–º–µ–Ω—ñ "Antonina" –±—É–¥–µ –≤–∏–≤–µ–¥–µ–Ω–æ –¥–≤–∞ –ø–æ–≤—ñ–¥–æ–º–ª–µ–Ω–Ω—è.

```python
name="Oksana"
separation= [char.lower() for char in name]
printed=False

if 'a' in separation or "i" in separation:
  print("A or I is in the name")
  printed=True
if "o" in separation and "n" in separation:
  print( "O and N is in the name")
  printed=True
if not "o" in separation:
   if "n" in separation:
    print("Only N is in the name")
    printed=True
if not printed:
  print( "There is no N in the name, but there might be O")
```

–ü—ñ–¥–∫–∞–∑–∫–∞: –≤–∏–∫–æ—Ä–∏—Å—Ç–æ–≤—É–π—Ç–µ —Ü–∏–∫–ª while —Ç–∞ —É–º–æ–≤–∏ if-else, —â–æ–± –ø–æ–≤—ñ–¥–æ–º–∏—Ç–∏ –∫–æ—Ä–∏—Å—Ç—É–≤–∞—á—É –ø—ñ–∫–∞–∑–∫–∏ (–Ω–∞–ø—Ä. "–ë—ñ–ª—å—à–µ", "–ú–µ–Ω—à–µ") –ø–æ–∫–∏ –∫–æ—Ä–∏—Å—Ç—É–≤–∞—á –Ω–µ –≤–≥–∞–¥–∞—î. –ü–æ –∑–∞–≤–µ—Ä—à–µ–Ω–Ω—é –≤–∏–≤–µ–¥—ñ—Ç—å –∫—ñ–ª—å–∫—ñ—Å—Ç—å —Å–ø—Ä–æ–±, —è–∫—ñ –∑–Ω–∞–¥–æ–±–∏–ª–∏—Å—è –¥–ª—è –≤–≥–∞–¥—É–≤–∞–Ω–Ω—è.

```python
import random
n=random.randint(1,100)
numb=int(input("Girl, enter your numbüòú : "))
if numb>100:
  print("–£—Å–ø–∞–∫–æ–π—Å—è, –¥–æ 100 —Ä–∞–∑—Ä—ñ—à–∞—é.")

while n!=numb:
  if n>numb:
    print("–ú–∞–ª–∏—à–∫–∞ —Å–ø—Ä–æ–±—É–π –±—Ä–∞—Ç–∏ –≤–∏—â–µüòú")
    if n-numb>90:
      print("—Ç–∏ —á–æ—Ç–∞ –≤–∑–∞–≥–∞–ª—ñ –Ω–µ —Ç—É–¥–∏ –ø—ñ—à–ª–∞ü´†")
  else:
    print("–ü–∞–ø—É—Å—Ç—ñ—Å—å")

  numb=int(input("–î–∞–≤–∞–π –∑—Ä–æ–±–∏–º –≤–∏–≥–ª—è–¥, —â–æ —Ç–æ–≥–æ –ø–æ–∑–æ—Ä—É –Ω–µ –±—É–ª–æüôÑ: "))

print("PERIODDD, –Ω–∞—Ä–µ—à—Ç—ñ –≤–∏–≥—Ä–∞–ª–∞ ")

```
